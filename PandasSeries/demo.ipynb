#### Creating a DataFrames

##### 1.Using a Dictionary

```python
from pandas import DataFrame


data = {
    "Name":["aaa","bbb","ccc","ddd"],
    "usn":[10,20,30,40],
    "age":[18,66,32,42]
}
df = pd.DataFrame(data)
df
```

#### 2. Using a List

```python
data = [
    ["AAA",10,45],
    ["BBB",20,22],
    ["CCC",30,23]
]
df = pd.DataFrame(data,columns=["Name","usn","age"])
df
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Husen", "Ali", "Sara", "Ayaan"],
    "Age": [22, 23, 21, 24]
}, index=[101, 102, 103, 104])
df
```

```python
df.loc[101]                    # Row with index label 101
```

```python
df.loc[101, "Name"]            # Husen
```

```python
df.loc[101:103, ["Name","Age"]]  # Inclusive slicing
```

```python
df.iloc[0:3, 0:2]
```
