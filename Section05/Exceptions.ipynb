#### Value Error

```python
try:
    age = int("abc")
    print(age)
except ValueError as ex:
    print("Value Error is Handled")
    print(ex)
```

#### TypeError

```python
try:
    num = "10" + 6
    print(num)
except TypeError as ex:
    print("Type Error is Handled")
    print(ex)
```

#### Index Error

```python
a = [10,50,30]
try:
    print(a[10])
except IndexError as ex:
    print("Type Error is Handled")
    print(ex)
```

#### Key Error

```python
person = {"name":"Husen","age":23}
try:
    print(person["city"])
except KeyError as ex:
    print("Key Error is Handled")
    print(ex)
```

#### File Handling Exception

```python
try:
    file=open("sample.txt","r")
except FileNotFoundError as ex:
    print("Handled")
    print(ex)
else:
    print("FIle is Present")

```
