##### Divide By Zero Exception

```python
num = 10
try:
    result = num/0
    print(result)
except ZeroDivisionError as ex:
    print(ex)
else:
    print("Program Run Successfully")
finally:
    sum = num + 10
    print(sum)
```

#### Value Error

```python
try:
    num = int("aaa")
except ValueError as ex:
    print(ex)
    print("Invalid values has been passed")
else:
    print("Program Run Successfully")
```

#### Type Error

```python
try:
    num = "10"+12
except TypeError as ex:
    print(ex)
else:
    print("Type Values are validated Successfully")
finally:
    print("Finally Block Exucuted Successfully...")
```

#### Key Error

```python
student = {
    "name":"Husen",
    "usn":123
}
try:
    print(student["city"])
except KeyError as ex:
    print(ex)
    print("Key is not present in a Dict")
else:
    print("Data Validated Successfully")
finally:
    print("Finally Block Exucuted Successfully...")
```

#### Index Error

```python
list = [100,200,600,300]
try:
    print(list[10])
except IndexError as ex:
    print(ex)
    print("Trying to access invalid index")
else:
    print("Elements are accessed Successfully")
finally:
    print("Finally Block Exucuted Successfully...")
```
