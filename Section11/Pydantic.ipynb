```python
from pydantic import BaseModel
from langgraph.graph import START,END,StateGraph

class State(BaseModel):
    name:str
```

### Create a Node

```python
def nodes(state:State):
    return {"name":"Tosif "}
```

### Create a Graph

```python
from IPython.display import Image,display

#Create a Graph
graph = StateGraph(State)

#Add Nodes
graph.add_node("node", nodes)

## Adding a Edges
graph.add_edge(START,"node")
graph.add_edge("node",END)

#Complie the Graph
g_builder = graph.compile()

display(Image(g_builder.get_graph().draw_mermaid_png()))
```

### Invoke The Graph

```python
g_builder.invoke({"name":""})
```
